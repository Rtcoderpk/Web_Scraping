```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.whatmobile.com.pk/35000_to_45001_Mobiles'
```

```python
website = requests.get(URL)
```

```python
website
```

```python
website = requests.get(URL).text
```

```python
website
```

```python
soup = BeautifulSoup(website, 'lxml')
```

```python
soup
```

```python
mobile_container = soup.find('td', class_ = 'BiggerText')
```

```python
mobile_container
```

```python
mobile_name = mobile_container.find('h4', class_ = 'p4 biggertext')
```

```python
mobile_name
```

```python
mobile_price = mobile_container.find('span' , class_ = 'PriceFont' ).text
```

```python
mobile_price
```

```python
mobile_container = soup.find_all('td', class_ = 'BiggerText')
for mobile in mobile_container:
  mobile_name = mobile.find('h4', class_ = 'p4 biggertext').text.replace('\n',"")
  mobile_price = mobile.find('span' , class_ = 'PriceFont' ).text.replace('\n',"")

  print(f'Mobile Name: {mobile_name}, Price: {mobile_price}')
```
